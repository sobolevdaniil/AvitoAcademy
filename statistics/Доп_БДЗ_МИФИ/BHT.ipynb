## –ó–∞–¥–∞—á–∞ 1. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ó–ë–ß

–ü–æ –æ–¥–Ω–æ–π –∏–∑ —Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–æ–∫ –∑–∞–∫–æ–Ω–∞ –±–æ–ª—å—à–∏—Ö —á–∏—Å–µ–ª, –¥–ª—è –°–í $\xi$ —Å $M(\xi) = ùëé < +\infty$ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ:
 $\langle ùúâùëõ \rangle \rightarrow ùëé$, –≥–¥–µ $\langle ùúâùëõ \rangle = \frac{1}{n} \sum_{k=1}^n \xi_k  $

–ü—É—Å—Ç—å: 

- $N$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤ —Å–ª—É—á–∞–π–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω
- $n$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –Ω–∞–±–æ—Ä–µ <br>
- c–∏–º–≤–æ–ª–æ–º $j$ –±—É–¥–µ–º –æ–±–æ–∑–Ω–∞—á–∞—Ç—å –Ω–æ–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏

```python
def plot(N: int, means: list, mean: float, label1: str, label2: str) -> None:
    """
        –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
        - N - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
        - means - —Å–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ –≤—ã–±–æ—Ä–æ–∫
        - mean - –∏—Å—Ç–∏–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–Ω–µ –∑–Ω–∞—á–µ–Ω–∏–µ       
    """
    plt.figure(figsize=(15, 30))
    plt.title('–û—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è', fontsize=12)
    plt.scatter(means, range(N), linewidth=2, label = label1)
    plt.plot([mean for i in range(N)], range(N), linestyle='--', color='g', label = label2)

    plt.xlabel('–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ', fontsize=12)
    plt.ylabel('–ù–æ–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏', fontsize=12)
    plt.legend(fontsize=12)
    plt.show()
```

```python
def density_plot(means: list) -> None:
    """
        –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
        - means - —Å–ø–∏—Å–æ–∫ —Å –∞–≥–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤—ã–±–æ—Ä–æ–∫
    """
    plt.figure(figsize=(10, 5))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π', fontsize=12)
    sns.distplot(means, color='b', label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
    
    plt.xlabel('X', fontsize=12)
    plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å', fontsize=12)
    plt.legend(fontsize=12)
    plt.show()
```

```python
def find_means(N: int, n: int, generator) -> list:
    """
        –§—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∏—Ç —Å–ø–∏—Å–æ–∫ —Å–æ —Å—Ä–µ–¥–Ω–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
        - N - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
        - n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–µ
    """
    means = []
    for j in range(N):
        collection = generator()
        mean = collection.mean()
        means.append(mean)
    return means
```

```python
def norm_generator(n: int) -> np.array:
    return np.random.randn(n)
```

```python
def expon_generator(L: float, n: int) -> np.array:
    return np.random.exponential(1/L, n)
```

```python
def poisson_generator(L: float, n: int) -> np.array:
    return np.random.poisson(1/L, n)
```

### –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ

- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ - –Ω–æ–ª—å
- –î–∏—Å–ø–µ—Ä—Å–∏—è - –µ–¥–µ–Ω–∏—Ü–∞

–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–æ–ª–∂–Ω–æ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –º–µ—Ç–º–∞—Ç–∏—á–µ—Å–∫–æ–º—É –æ–∂–∏–¥–∞–Ω–∏—é, —Ç–æ –µ—Å—Ç—å –∫ –Ω—É–ª—é

```python
means = find_means(400, 10_000, lambda: norm_generator(10_000))
plot(400, means, 0, '–°—Ä–µ–¥–Ω–Ω–µ j-–æ–π –≤—ã–±–æ—Ä–∫–∏', '–ú–∞—Ç –æ–∂–∏–¥–∞–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
```

```python
density_plot(means)
```

–í —Ç–µ–æ—Ä–∏–∏ —á–µ–º –±–æ–ª—å—à–µ –≤—ã–±–æ—Ä–∫–∞, —Ç–µ–º –º–µ–Ω—å—à–µ –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ.

```python
def convergence_plot_norm(n_max: int, true_mean: float) -> None:
    """
        –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –≤—ã–±–æ—Ä–∫–µ 
        –æ—Ç –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—ã–±–æ—Ä–∫–∏ –æ—Ç –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –ø—Ä–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏
        
        - n - –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ
        - true_mean - –∏—Å—Ç–∏–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ
    """
    plt.figure(figsize=(20, 20))
    
    for n in range(1, n_max):
        collection = norm_generator(n)
        mean = collection.mean()
        plt.plot(n, mean, '_', alpha = 1, linewidth=2)
        
    plt.plot(range(n), [true_mean for i in range(n)], linestyle='--', color='b', label = '–ò—Å—Ç–∏–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ', linewidth=3)

    plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –≤—ã–±–æ—Ä–∫–µ', fontsize=12)
    plt.ylabel('–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ', fontsize=12)
    plt.ylim([true_mean - 0.3, true_mean + 0.3])
    plt.legend(fontsize=12)
    plt.show()
```

```python
convergence_plot_norm(10_000, 0)
```

## –ó–∞–¥–∞—á–∞ 2. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ó–∞–∫–æ–Ω–∞ –£—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏ –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö –ß–∞—Å—Ç–æ—Ç.

–ü—É—Å—Ç—å $\mu_n(A) = \frac {n(A)} {n}$ - –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è —á–∞—Å—Ç–æ—Ç–∞ –Ω–∞—Å—Ç—É–ø–ª–µ–Ω–∏—è —Å–æ–±—ã—Ç–∏—è A –≤ n –æ–ø—ã—Ç–∞—Ö, –≥–¥–µ $A(n)$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø—ã—Ç–æ–≤, –≤ –∫–æ—Ç–æ—Ä—ã—Ö —Å–æ–±—ã—Ç–∏–µ –ê –Ω–∞—Å—Ç—É–ø–∏–ª–æ. –¢–æ–≥–¥–∞ $\mu_n \rightarrow P(A)$

–í –∫–∞—á–µ—Å—Ç–≤–µ–µ —Å–æ–±—ã—Ç–∏—è –ê:
- –í—ã–ø–∞–¥–µ–Ω–∏–µ 7 –æ—á–∫–æ–≤ –ø—Ä–∏ –±—Ä–æ—Å–∞–Ω–∏–∏ 2-—É—Ö –∫–æ—Å—Ç–µ–π
- –í—ã—Ç—è–≥–∏–≤–∞–Ω–∏–µ 5 –∫–∞—Ä—Ç –∏–∑ –∫–æ–ª–æ–¥—ã 52 –∫–∞—Ä—Ç—ã (4 –º–∞—Å—Ç–∏, 13 –Ω–æ–º–∏–Ω–∞–ª–æ–≤) —Ç–∞–∫, —á—Ç–æ —Å—Ä–µ–¥–∏ –∏—Ö –æ–∫–∞–∂—É—Ç—Å—è –¥–≤–∞ —Ç—É–∑–∞, –¥–≤–∞ –∫–æ—Ä–æ–ª—è –∏ –æ–¥–Ω–∞ —Å–µ–º–µ—Ä–∫–∞.

–ü—É—Å—Ç—å: 
- $N$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–æ—Ä–æ–≤, $j$ - –Ω–æ–º–µ—Ä –Ω–∞–±–æ—Ä–∞ 
- $n$ - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–∞—Ä–Ω—ã—Ö –∏—Å—Ö–æ–¥–æ–≤ –≤ –∫–∞–∂–¥–æ–º –Ω–∞–±–æ—Ä–µ

### –ë—Ä–æ—Å–∞–Ω–∏–µ –∫—É–±–∏–∫–æ–≤

–ù–∞–π–¥–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–µ–Ω–∏–µ $P(A)$:

$P(A) = \frac m n$, –≥–¥–µ
- m - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã—Ö –∏—Å—Ö–æ–¥–æ–≤
- n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Å–µ–≤–æ–∑–º–æ–∂–Ω—ã—Ö –∏—Å—Ö–æ–¥–æ–≤

–ë–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã–µ –∏—Å—Ö–æ–¥—ã: (1,6) (6,1) (2,5) (5,2) (3,4) (4,3) <br>
–ò—Ö –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ: 6

–í—Å–µ–≤–æ–∑–º–æ–∂–Ω—ã–µ –∏—Å—Ö–æ–¥—ã: –∑–Ω–∞—á–µ–Ω–∏–µ –Ω–∞ –∫–∞–∂–¥–æ–º –∫—É–±–∏–∫–µ –≤—ã–ø–∞–¥–∞–µ—Ç –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –¥—Ä—É–≥ –¥—Ä—É–≥–∞, —Ç–æ –µ—Å—Ç—å –≤—Å–µ–≥–æ –∏—Å—Ö–æ–¥–æ–≤ 6*6

–¢–æ–≥–¥–∞ $P(A) = \frac 6 {36} = \frac 1 6 \approx 0.167$ 

```python
def find_selection(N: int, n: int, point: int) -> list:
    """
        –§—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Å–æ–±—ã—Ç–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–∏
        - N - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
        - n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –≤—ã–±–æ—Ä–∫–µ
        - point - —Å–æ–±—ã—Ç–∏–µ
    """
    selection = []
    for j in range(N):
        count = 0
        for i in range(n):
            a = np.random.randint(1, 7)
            b = np.random.randint(1, 7)
            if (a + b) == point:
                count += 1
        selection.append(count/n)
    return selection
```

```python
selection = find_selection(400, 500, 7)
plot(400, selection, 1/6, '–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤ j-–æ–π –≤—ã–±–æ—Ä–∫–µ', '–ò—Å—Ç–∏–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
```

```python
density_plot(selection)
```

### –í—ã—Ç—è–≥–∏–≤–∞–Ω–∏–µ –∫–∞—Ä—Ç

–ù–∞–π–¥–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–µ–Ω–∏–µ $P(A)$:

$P(A) = \frac m n$, –≥–¥–µ
- m - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã—Ö –∏—Å—Ö–æ–¥–æ–≤
- n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Å–µ–≤–æ–∑–º–æ–∂–Ω—ã—Ö –∏—Å—Ö–æ–¥–æ–≤

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ –∏–∑–≤–ª–µ—á—å –¥–≤–∞ —Ç—É–∑–∞: $C_{4}^{2} = \frac {4!} {2!2!} = 6$ <br>
–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ –∏–∑–≤–ª–µ—á—å –¥–≤–∞ –∫–æ—Ä–æ–ª—è: $C_{4}^{2} = \frac {4!} {2!2!} = 6$ <br>
–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ –∏–∑–≤–ª–µ—á—å –æ–¥–Ω—É —Å–µ–º–ø–µ—Ä–∫—É: $C_{4}^{1} = \frac {4!} {3!1!} = 4$

–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–æ—Å–æ–±–æ–≤ –∏–∑–≤–ª–µ—á—å –ø—è—Ç—å –∫–∞—Ä—Ç –∏–∑ –ø—è—Ç–∏–¥–µ—Å—è—Ç–∏ –¥–≤—É—Ö: $C_{52}^{5} = \frac {52!} {47!5!} = \alpha$ 

–¢–æ–≥–¥–∞ $P(A) = \frac {6*6*4} \alpha$

- –í—Å–µ–≥–æ —Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç—å –Ω–æ–º–∏–Ω–∞–ª–æ–≤, –ø—É—Å—Ç—å —á–∏—Å–ª–∞ –æ—Ç –æ–¥–Ω–æ–≥–æ –¥–æ —Ç—Ä–∏–Ω–∞–¥—Ü–∞—Ç–∏ –æ—Ç–≤–µ—á–∞—é—Ç –∏–º
- –¢–æ–≥–¥–∞ –Ω–∞–º –Ω—É–∂–Ω–æ: –¥–≤–∞ —Ä–∞–∑–∞ 13, –¥–≤–∞ —Ä–∞–∑–∞ 12 –∏ –æ–¥–∏–Ω —Ä–∞–∑ 6

```python
def find_collection(N: int, n: int, points: list) -> list:
    """
        –§—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Å–æ–±—ã—Ç–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–∏
        - N - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
        - n - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –≤—ã–±–æ—Ä–∫–µ
        - points - —Å–ø–∏—Å–æ–∫ —Å –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
    """
    collection = []
    for j in tqdm(range(N)):
        count = 0 
        for i in range(n):
            cards = [
                np.random.randint(1, 14),
                np.random.randint(1, 14),
                np.random.randint(1, 14),
                np.random.randint(1, 14),
                np.random.randint(1, 14),
            ]
            
            for point in points:
                if point in cards:
                    cards.remove(point)
            if cards == []:
                count += 1
        collection.append(count/n)
    return collection
```

```python
collection = find_collection(400, 100_000, [13, 13, 12, 12, 6])
```

- –ü–æ –≥—Ä–∞—Ñ–∏–∫—É –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –º–æ–∂–Ω–æ –ø–æ–¥—É–º–∞—Ç—å, —á—Ç–æ –≤—Å—ë –æ–∫, —Ç–∞–∫ –∫–∞–∫ –≤–µ–ª–∏—á–∏–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏–ª–∞—Å—å –±–ª–∏–∑–∫–∞ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É. –ù–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –≤–æ–∫—Ä—É–≥ –Ω–µ –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏! –ê –º–æ–∂–µ—Ç —è –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ –≤—ã—Å—á–∏—Ç–∞–ª –∏—Å—Ç–∏–Ω–Ω–æ–µ :)

## –ó–∞–¥–∞—á–∞ 3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –∫–ª. –¶–ü–¢.

–ü—É—Å—Ç—å $\xi_1, ..., \xi_n$ &mdash; **–Ω–µ–∑–∞–≤–∏—Å–∏–º–æ** –æ–¥–∏–Ω–∞–∫–æ–≤–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–µ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã, —É –∫–æ—Ç–æ—Ä—ã—Ö —Å—É—â–µ—Å—Ç—É—é—Ç –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è: $E [\xi_i] = \mu < \infty$ –∏ $Var[\xi_i] = \sigma^2 < \infty$. 
–¢–æ–≥–¥–∞ $\sqrt{n}\dfrac{\overline \xi - \mu}{\sqrt{\sigma^2}}$ [—Å—Ö–æ–¥–∏—Ç—Å—è –ø–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é](https://en.wikipedia.org/wiki/Convergence_in_distribution) –∫ $\mathcal{N}(0, 1)$, 
–≥–¥–µ $\overline \xi -$ —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≤—ã–±–æ—Ä–∫–∏

### –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ:

–ü—Ä–∏–º–µ–º –ø–∞—Ä–º–µ—Ç—Ä $\lambda$ —Ä–∞–≤–Ω—ã–π 10, —Ç–æ–≥–¥–∞:

- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ - 1/$\lambda$ = 1/10
- –î–∏—Å–ø–µ—Ä—Å–∏—è - 1/$\lambda^2$ = 1/100

```python
def find_clt_values(N:int, sample_generator, expected_value:float, variance:float) -> list:
    """
        –§—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∞–µ—Ç —Å–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¶–ü–¢
        –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
            - sample_generator: —Ñ—É–Ω–∫—Ü–∏—è-–≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –≤—ã–±–æ—Ä–∫–∏. –ù–µ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∏—á–µ–≥–æ –Ω–∞ –≤—Ö–æ–¥
            - expected_value: –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏
            - variance: –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã–±–æ—Ä–∫–∏
    """
    clt_sample = []
    for _ in range(N):
        # –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º N —Ä–∞–∑ –≤—ã–±–æ—Ä–∫—É
        sample = sample_generator()

        # —Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –∏–∑ –¶–ü–¢
        sample_size = len(sample)
        statistic = np.sqrt(sample_size) * (np.mean(sample) - expected_value) / np.sqrt(variance)

        #—Å–æ—Ö—Ä–∞–Ω—è–µ–º
        clt_sample.append(statistic)
    return clt_sample
```

```python
def visualize_relation(N:int, clt_sample: list, label1: str) -> None:
    """
        –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
        - N - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
        - means - —Å–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ –≤—ã–±–æ—Ä–æ–∫
        - mean - –∏—Å—Ç–∏–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–Ω–µ –∑–Ω–∞—á–µ–Ω–∏–µ       
    """
    plt.figure(figsize=(15, 30))
    plt.title('–û—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è', fontsize=12)
    plt.scatter(clt_sample, range(N), linewidth=2, label = label1)

    plt.xlabel('–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞', fontsize=12)
    plt.ylabel('–ù–æ–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏', fontsize=12)
    plt.legend(fontsize=12)
    plt.show()
```

```python
def visualize_CLT(clt_sample: np.array) -> None:
    """
        –§—É–Ω–∫—Ü–∏—è-–≤–∏–∑—É–∞–ª–∏–∑–∞—Ç–æ—Ä –¶–ü–¢. –°—Ç—Ä–æ–∏—Ç:
            - –≥–∏—Å—Ç–æ—Ä–≥–∞–º–º—É —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¶–ü–¢,
            - —ç–º–ø–∏—Ä–∏—á–µ—Å–∫—É—é –ø–ª–æ—Ç–Ω–æ—Å—Ç—å,
            - –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
        - clt_sample - –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¶–ü–¢.
    """

    x = np.linspace(-4, 4, 1000)
    plt.figure(figsize=(10, 5))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π', fontsize=12)
    sns.distplot(clt_sample, color = 'b', label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
    plt.plot(x, norm().pdf(x), label='$\mathcal{N}(0, 1)$')
    
    plt.xlabel('X', fontsize=12)
    plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å', fontsize=12)
    plt.legend(fontsize=12)
    plt.show()
```

–°–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã –º–æ–≥—É—Ç –±—ã—Ç—å —Å–ª–∞–±–æ –∑–∞–≤–∏—Å–∏–º—ã –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞ –∏ —Å–ª–µ–≥–∫–∞ –ø–æ-—Ä–∞–∑–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã. –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω–∞—è –ø—Ä–µ–¥–µ–ª–µ–Ω—å–Ω–∞—è —Ç–µ–æ—Ä–º–∞ –≤—Å–µ –µ—â–µ –±—É–¥–µ—Ç [–≤–µ—Ä–Ω–∞](https://en.wikipedia.org/wiki/Stable_distribution#A_generalized_central_limit_theorem).

## –ó–∞–¥–∞—á–∞ 4. –≠–ª–µ–º–µ–Ω—Ç—ã –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏.

–û—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.


-  –í –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–π —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–µ –æ—Ü–µ–Ω–∫–æ–π –Ω–µ–∫–æ—Ç–æ—Ä–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $\theta$ —Å–≤—è–∑–∞–Ω–Ω–æ–≥–æ —Å–æ —Å–ª—É—á–∞–π–Ω–æ–π  –≤—ã–±–æ—Ä–∫–æ–π $(X_1, X_2, ..., X_n)$, –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ—Ç –≤—ã–±–æ—Ä–∫–∏ $\widehat \theta = \widehat \theta_n = \theta(ùëã_1, ùëã_2, ... , ùëã_n).$
- –û—Ü–µ–Ω–∫–∞ –ø—Ä–∞–º–µ—Ç—Ä–∞ $\theta$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è <u>–Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π</u>, –µ—Å–ª–∏ $M\widehat\theta_n = \theta$
- –û—Ü–µ–Ω–∫–∞ $\theta$ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è <u>—Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ–π</u>, –µ—Å–ª–∏ $\forall \varepsilon > 0 \lim\limits_{n \to +\infty} \{P|\widehat\theta_n - \theta| < \varepsilon\} = 1$, <br>
—Ç–æ –µ—Å—Ç—å –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –±–æ–ª—å—à–∏—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –æ—Ü–µ–Ω–∫–∏ –æ—Ç –æ—Ü–µ–Ω–∏–≤–∞–µ–º–æ–≥–æ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –Ω—É–ª—é.

–°–æ–≥–ª–∞—Å–Ω–æ –º–µ—Ç–æ–¥—É –º–æ–º–µ–Ω—Ç–æ–≤, –µ—Å–ª–∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è –æ—Ü–µ–Ω–∏—Ç—å –≤–µ–ª–∏—á–∏–Ω—É $a_j = MX_{1}^{j}$ - –º–æ–º–µ–Ω—Ç –ø–æ—Ä—è–¥–∫–∞ j, —Ç–æ –∫–∞—á–µ—Å—Ç–≤–µ –æ—Ü–µ–Ω–∫–∏ —Å–ª–µ–¥—É–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω—ã–π –º–æ–µ–Ω—Ç –ø–æ—Ä—è–¥–∫–∞ j:

$$\widehat a_j = \frac 1 n \sum_{k = 1}^{n} X_{k}^{j}$$

–ï—Å–ª–∏ –∂–µ –Ω—É–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –≤–µ–ª–∏—á–∏–Ω—É $b_j = M(X_1 - MX_1)^j$ - —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π –º–æ–º–µ–Ω—Ç –ø–æ—Ä—è–¥–∫–∞ j, —Ç–æ –æ—Ü–µ–Ω–∫–æ–π —Å–ª—É–∂–∏—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π –≤—ã–±–æ—Ä–æ—á–Ω—ã–π –º–æ–º–µ–Ω—Ç –ø–æ—Ä—è–¥–∫–∞ j, –∑–∞–¥–∞–≤–∞–µ–º—ã–π —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ–º:
$$\widehat b_j = \frac 1 n \sum_{k=1}^n (X_k - \overline X)^j,$$
–≥–¥–µ $\overline X$ - —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –≤—ã–±–æ—Ä–∫–∏

- –í—ã–±–æ—Ä–æ—á–Ω—ã–π –º–æ–º–µ–Ω—Ç $\widehat a_j$ —è–≤–ª—è–µ—Ç—Å—è <u> –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –∏ —Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π.</u>
- –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–π –≤—ã–±–æ—Ä–æ—á–Ω—ã–π –º–æ–º–µ–Ω—Ç $\widehat b_j$ –¥–ª—è $j \geq 2$ —è–≤–ª—è–µ—Ç—Å—è <u>—Å–º–µ—â–µ–Ω–Ω–æ–π</u> –∏ –ø—Ä–∏ —ç—Ç–æ–º <u>—Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π.</u>
- –î–ª—è –æ—Ü–µ–Ω–∫–∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –∏—Å–ø–æ–ª—å–∑—É—é—Ç –≤—ã–±–æ—Ä–æ—á–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç $\widehat a_1$
- –î–ª—è –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ $\sigma^2 = b_2$
    - –ü—Ä–∏ –º–∞–ª—ã—Ö $n$ –∏—Å–ø–æ–ª—å–∑—É—é—Ç –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É $S^2 = \frac {n} {n-1} \widehat\sigma^2 = \frac n {n-1} \widehat b_2$ 
    - –ü—Ä–∏ –±–æ–ª—å—à–∏—Ö $n$ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω—ã–π $\widehat \sigma^2 = \widehat b_2$, —Ç–∞–∫ –∫–∞–∫ –µ—ë —Å–º–µ—â–µ–Ω–∏–µ –Ω–µ—ã–µ–ª–∏–∫–æ.

<u>**–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–≥–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:**</u> - —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞, –æ–ø—Ä–µ–¥–µ–ª—è–µ–º–∞—è —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ–º:

$$F_n^*(x) = \frac 1 n \sum_{k=1}^{n}\chi \{X_k < x\} = \frac 1 n \nu_n(x),$$

–≥–¥–µ $\nu_n(x) =$ (—á–∏—Å–ª–æ —Ç–µ—Ö $X_k$ —Å—Ä–µ–¥–∏ $X_1, X_2, X_3 ..., X_n$, –∫–æ—Ç–æ—Ä—ã–µ –º–µ–Ω—å—à–µ $x)$;
$\chi(A)$ - –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä —Å–æ–±—ã—Ç–∏—è. 

–ú–æ–∂–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ $F_n^*(x)$ —è–≤–ª—è–µ—Ç—Å—è <u>–Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –∏ —Å–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ–π</u> –æ—Ü–µ–Ω–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $F(x)$ –≤ –∫–∞–∂–¥–æ–π –æ—á–∫–µ $x \in (-\infty;\infty).$

### –û—Ü–µ–Ω–∫–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è

–ù–∞–º –Ω–∞–¥–æ –æ—Ü–µ–Ω–∏—Ç—å $a_j = MX_{1}^{j}$ - –º–æ–º–µ–Ω—Ç –ø–æ—Ä—è–¥–∫–∞ $j$, –ø—Ä–∏ $j = 1$, —Ç–æ–≥–¥–∞ –≤–æ—Å–ø–æ–ª—å–∑—É–µ–º—Å—è —Å–ª–µ–¥—É—é—â–µ–π —Ñ–æ—Ä–º—É–ª–æ–π:
$$\widehat a_1 = \frac 1 n \sum_{k = 1}^{n} X_{k}$$

```python
def plot_sample_means(data, size:int, steps:int) -> None:
    """
        –§—É–Ω–∫—Ü–∏—è —Å—Ç—Ä–æ–∏—Ç –≥—Ä–∞—Ñ–∏–∫ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        - steps - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—É—á–∞–µ–º—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö
        - size - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫–æ–∂–¥–æ–π –ø–æ–¥–≤—ã–±–æ—Ä–∫–µ
    """
    
    means = [data.sample(size).mean() for i in range(steps)]
    
    plt.figure(figsize=(10, 5))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π', fontsize=12)
    sns.distplot(means, color='b', label='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π')
    
    plt.xlabel('–°—Ä–µ–¥–Ω–∏–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ', fontsize=12)
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ', fontsize=12)
    plt.legend(fontsize=12)
    plt.show()                          
```

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –≥—Ä–∞—Ñ–∏–∫ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö. –í —Ç–µ–æ—Ä–∏–∏ –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ —Å—Ä–µ–¥–Ω–∏–µ –¥–æ–ª–∂–Ω—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–∏—Ç—å—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ –≤–æ–∫—Ä—É–≥ –∏—Å—Ç–∏–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ:

```python
plot_sample_means(data, 30, 1000)
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–∞ –≤–∏–¥–Ω–æ, —á—Ç–æ –∏—Å—Ç–∏–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –¥–æ–ª–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—Å—è —á—É—Ç—å –±–æ–ª—å—à–µ 30 000

```python
def expectation_estimate(data, size, steps) -> float:
    """
        –§—É–Ω–∫—Ü–∏—è –Ω–∞–π–¥–µ—Ç –æ—Ü–µ–Ω–∫—É –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è
    """
    return data.mean()
```

```python
print(f'–û—Ü–µ–Ω–∫–∞ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è: {expectation_estimate(data, 30, 1000)}')
```

### –û—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏

–ù–∞–π–¥–µ–º –∫–∞–∫ —Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É, —Ç–∞–∫ –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é.

- –ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è:

$$S^2 = \frac {n} {n-1} \widehat\sigma^2 = \frac n {n-1} \widehat b_2 = \frac n {n-1} \frac 1 n \sum_{k=1}^n (X_k - \overline X)^2 = \frac 1 {n-1}  \sum_{k=1}^n (X_k - \overline X)^2$$

- –°–º–µ—â–µ–Ω–Ω–∞—è:

$$\widehat\sigma^2 = \frac 1 n \sum_{k=1}^n (X_k - \overline X)^2$$

```python
def dispersion_estimate(data, size:int, steps:int) -> float:
    """
        –§—É–Ω–∫—Ü–∏—è –Ω–∞–π–¥–µ—Ç –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏ 
    """
    return np.var(data)
```

```python
print(f'–°–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏: {dispersion_estimate(data, 30, 1000)}')
print(f'–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏: {dispersion_estimate(data, 30, 1000) * 30 / 29}')
```

### –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (–±–µ–∑ —É—á–µ—Ç–∞ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏)

<u>–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–≥–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:</u>

$$F_n^*(x) = \frac 1 n \sum_{k=1}^{n}\chi \{X_k < x\} = \frac 1 n \nu_n(x),$$

–≥–¥–µ $\nu_n(x) =$ (—á–∏—Å–ª–æ —Ç–µ—Ö $X_k$ —Å—Ä–µ–¥–∏ $X_1, X_2, X_3 ..., X_n$, –∫–æ—Ç–æ—Ä—ã–µ –º–µ–Ω—å—à–µ $x)$;
$\chi(A)$ - –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä —Å–æ–±—ã—Ç–∏—è. 

```python
def plot_empirical_distribution_function(data):
    plt.figure(figsize=(10, 10))
    plt.title('–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', fontsize=12)
    plt.hist(data, histtype='step', color='b', cumulative=True, bins=len(data))
    plt.xlim([0, 200_000])
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ x', fontsize=12)
    plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ $F(x)$', fontsize=12)
    plt.show()
```

```python
plot_empirical_distribution_function(data)
```

### –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –∫—Ä–∏–≤—ã—Ö –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ

–ü–æ—Å—Ç—Ä–æ–∏–º –Ω–∞ –æ–¥–Ω–æ–º –≥—Ä–∞—Ñ–∏–∫–µ —Å–ª–µ–¥—É—é—â–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:
- –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –ø–æ –∑–Ω–∞—á–µ–Ω–∏—è–º –≤—ã–±–æ—Ä–∫–∏
- –ü–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
- –ü–ª–æ—Ç–Ω–æ—Å—Ç—å –ü—É–∞—Å—Å–æ–Ω–æ–≤—Å–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
- –ü–ª–æ—Ç–Ω–æ—Å—Ç—å —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

–í –∫–∞—á–µ—Å—Ç–≤–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–æ–∑—å–º–µ–º –Ω–∞–π–¥–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏

```python
def distribution_chart(data):
    mean = expectation_estimate(data, 30, 1000)
    var = dispersion_estimate(data, 30, 1000) ** (1/2)
    
    x = np.linspace(-50_000, 300_000, 1_000_000)
    plt.figure(figsize=(10, 5))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π', fontsize=12)
    sns.distplot(data, color='b', label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')    
    plt.plot(x, norm.pdf(x, mean, var), label='–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
    plt.plot(x, expon.pdf(x, scale=mean), label='–≠–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ')
    
    plt.xlabel('X', fontsize=12)
    plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å', fontsize=12)
    plt.legend(fontsize=12)  
    
    plt.show()
```

```python
distribution_chart(data)
```
