Далее мы можем рассчитать эмперическое значение критерия - $\chi^2_{эмп} = \sum^n_{i=1}(X_{1} - X_{2})^2 / X_{2}$

## Задача 2 (4 балла)

Реализовать критерий хи-квадрат для таблиц сопряженности $N \cdot M$.

- Условия применимости прверять не надо,  критерий будет проверяться на выборках больших размеров. 

Какими библиотеками вы можете пользоваться:

````python
import numpy as np
from collections import namedtuple
from scipy.stats import chi2
````

````python
#Автопроверка

MyСhi2Results = namedtuple('MyСhi2Results', 
                                  ['is_rejected', 'pvalue'])


def my_chi2_for_contingency_tables(sample_A: list, sample_B: list, alpha: float = 0.05):
    """
    Параметры:
    - sample_A: первая выборка
    - sample_B: вторая выборка
    - alpha: уровень значимости критерия.
        
    Возвращает:
    - MyСhi2Results с полями:
        - is_rejected: bool
            - отверглась или нет гипотеза H_0 на уровне значимости alpha
        - pvalue: float
    """
    is_rejected = None
    pvalue = None
    
    k = 5 # корзины для выборки sample_A
    m = 5 # корзины для выборки sample_B
    
    # Разобъем sample_A и sample_B на k, m корзин соответсвенно, 
    # диапазоны которых будут равны между собой и протягиваться от мин до макс значения
    A = sorted(sample_A)
    B = sorted(sample_B)
    
    # считаем диапазоны для sample_A
    step_A = (A[-1] - A[0]) / k
    delta_A = [A[0]]
    for i in range(1, k):
        delta_A.append(delta_A[-1] + step_A)
    delta_A.append(A[-1])
    
    # считаем диапазоны для sample_B
    step_B = (B[-1] - B[0]) / m
    delta_B = [B[0]]
    for i in range(1, k):
        delta_B.append(delta_B[-1] + step_B)
    delta_B.append(B[-1])
    
    # считаем частоту попадания в каждый из диапазонов для sample_A
    frequency_A = [0]*k
    for v in A:
        for ind, delta in enumerate(delta_A[1:]):
            if v <= delta:
                frequency_A[ind] += 1
                break
                
    # считаем частоту попадания в каждый из диапазонов для sample_B            
    frequency_B = [0]*m
    for v in B:
        for ind, delta in enumerate(delta_B[1:]):
            if v <= delta:
                frequency_B[ind] += 1
                break
    
    AB = []
    for i in range(len(A)):
        AB.append((sample_A[i], sample_B[i]))
    
    frequencies = [[0] * m for i in range(k)]
    
    # считаем совместную частоту попадания
    for a, b in AB:
        for ind, delta in enumerate(delta_A[1:]):
            if a <= delta:
                ind_A = ind
                break
        for ind, delta in enumerate(delta_B[1:]):
            if b <= delta:
                ind_B = ind
                break
        frequencies[ind_A][ind_B] += 1
    
    n = len(sample_A)
    
    # Считаем статистику
    tau = 0
    for ind_A, line in enumerate(frequencies):
        for ind_B, v in enumerate(line):
            term = ((v - (frequency_A[ind_A]*frequency_B[ind_B])/n)**2) / (frequency_A[ind_A]*frequency_B[ind_B])
            tau += term
    tau *= n
    
    pvalue = 1 - chi2.cdf(x=tau, df=(m-1)*(k-1))
    if pvalue < alpha:
        is_rejected = True
    else:
        is_rejected = False

    return frequencies, MyСhi2Results(is_rejected, pvalue)
````

Известно, что 52 запроса дошли с первой попытки, 31 &mdash; со второй, 3 &mdash;  с третьей, а для доставки еще 5 запросов потребовалось более 4  попыток. Верно ли, что сервер получает запросы с вероятностью 2/3?

**Задачу нужно решить теоретически!**

Можно воспользоваться лишь библиотекой `from scipy.stats import chi2`.

Сервер получает запросы с вероятностью 2/3, соответственно не получает запрос с вероятностью 1/3. Тогда веротяность, что запрос дойдет:
- С первого раза - $2/3$
- Со второго раза - $2/3 * 1/3$
- С третьего раза - $2/3 * 1/3 * 1/3$
- С четвертого раза и более - $1 - (2/3 + 2/3 * 1/3 + 2/3 * 1/3 * 1/3)$

При уровне значимости 0.05 проверим следующую гипотезу:

$\textbf{H_0:}$ сервер получает запросы с вероятностью 2/3<br>
$\textbf{H_1:}$ $H_0$ неверна

````python
pd.DataFrame(
    [[52, 31, 3, 5], ['2/3*91', '2/9*91', '2/27*91', '1/27*91']], 
     index=['Эмпирическая частота', 'Теоретическая частота'], 
     columns=['С первого раза', 'Со второго раза', 'С третьего раза', 'Более чем с тертьего'],
)
````

Мы видим проблему - частота ячеек не везде больше 5, поэтому стоит объеденить ячейки запросов дошедших с третьего раз и более, чем с третьего.

````python
pd.DataFrame(
    [[52, 31, 8], ['2/3*91', '2/9*91', '1/9*91']], 
     index=['Эмпирическая частота', 'Теоретическая частота'], 
     columns=['С первого раза', 'Со второго раза', 'С третьего разаи более'],
)
````

Тогда нам нужно получить значение хи-квадрат $\chi^2_{} = \sum^n_{i=1}(X_{эмп} - X_{теор})^2 / X_{теор}$

И далее сравнить полученное значение со значением хи-квадрат для 2 степеней свобод, при уровне значимости 0.05

````python
from scipy.stats import chi2

1 - chi2.cdf(x=(52 - 2*91/3)**2 / (2*91/3) + (31 - 2*91/9)**2 / (2*91/9) + (8 - 91/9)**2 / (91/9), df=3 - 1)
````

**Вывод:**  pvalue < alpha = 0.05, следовательно, мы склоняемся в сторону альтернативной гипотезы: сервер получает запросы с иной вероятностью, не равной 2/3. О выполнении гипотезы  $𝐻_0$ ничего сказать нельзя.

(Можно отметить, что pvalue около 0.025, что не сильно близко к 0)

Компания по производству коромов для попугаев провела опрос среди 5000 владельцев попугаев о том, что предпочитают есть их питомцы. По результатам опроса оказалось, что 1010 птиц предпочитают глазированные сырки, 2200 &mdash; фрукты, 950 &mdash; семечки и крупы, а все остальные предпочитают колбасу. Компания сделала предположение, что вероятности предпочтения этих видов еды равны $\theta, 2\theta, \theta, 1-4\theta$, где $\theta \in (0, 1/4)$. Верно ли данное предположение?

**Задачу нужно решить теоретически!**

Можно воспользоваться лишь библиотекой `from scipy.stats import chisquare`.

P.S. Задача появилась из рациона попугая составителя домашнего задания...

При уровне значимости 0.05 проверим следующую гипотезу:

$\textbf{H_0:}$ Вероятности предпочтения видов равны $\theta, 2\theta, \theta, 1-4\theta$, где $\theta \in (0, 1/4)$<br>
$\textbf{H_1:}$ $H_0$ неверна

````python
pd.DataFrame(
    [[1010, 2200, 950, 840], ['$\theta$', '$2\theta$', '$\theta$', '1-4*$\theta$']], 
     index=['Эмпирическая частота', 'Теоретическая частота / 5000'], 
     columns=['Глазированные сырки', 'Фрукты', 'Семечки и крупы', 'Колбаса'],
)
````

Для того, чтобы критерий хи-квадрат работал на задаче с оценкой параметров, нам нужно оценить параметр с помощью метода максимального правдоподобия.

$P = \theta^{1010} * (2\theta)^{2200} * \theta^{950} * (1-4\theta)^{840}$

$ln(P) = ln(\theta)*1010 + ln(2)*2200 + ln(\theta)* 2200 + ln(\theta)*950 + ln(1-4\theta)*840$

Далее возьмем производную и приравняем к нулю:

$(1010 + 2200 + 950) * (1/\theta) + 840 * (1/(1-4\theta))*(-4) = 0$

$4160 * (1-4\theta) = 3360 * \theta$

Получим $\theta = 4160 / (3360 + 4160*4) = 0.208$ Данное значение попадает в предполагаемый интервал.

Так как мы оценили один параметр, следовательно, будем смотреть на хи-квадрат со степенями свобод: (4 - 1 - 1) (ddof=1)

````python
from scipy.stats import chisquare

chisquare([1010, 2200, 950, 840], f_exp=[0.208*5000, 2*0.208*5000, 0.208*5000, 0.168*5000], ddof=1)
````

**Вывод:** pvalue < alpha = 0.05, следовательно, основания принять нулевую гипотезу нет, склоняемся в сторону альтернативной.

Критерий однородности, который служит для проверки гипотезы о том, что выбрки сэмплированы из одного и того же распределения.

**Критерий Колмогорова-Смирнова**

$D_{m,n}$- статистика — максимальное расхождение мажду эмпирическими функциями распределения двух случайных величин
$$D_{m,n}= \sup_x|F_B(x) - F_A(x)|$$

Если эта разность значимо велика, распределения считаются различными.

Можно использовать только в случае непрерывных распределений.

**Двухвыборочный критерий Пирсона или Хи-квадрат**

Универсальный критерий для проверки гипотезы о различии распределений двух и более совокупностей.

Есть порог на минимальное количество наблюдаемых значений для каждой категории признака.

**Некий традиционный метод проверки однородности (Например, Критерий Стьюдента + Критерий Фишера)**

Используем t-test для проверки гипотезы о средних<br>
Используем f-test на равенство дисперсий

Если оба подтверждают гипотезу о равентсве, то можно говорить об однородности выборок

**Критерий знаков**
[Можно посмотреть тут](http://datalearning.ru/index.php/textbook?cid=1&mid=4&topic=3)

Из принятия основной гипотезы критерия знаков не следует однородность выборок, а следует лишь возможность однородности. Если же основная гипотеза критерия знаков отклоняется, то отклоняется и гипотеза однородности выборок.

От вас ожидаетя практический ответ на вопрос: будете ли вы в работе использовать этот критерий в качестве критерия однородности? Рассмотрите разные случаи и выборки. 

Сделайте финальный выбор.

Наверное, как неопытный статистик/аналитик можно тратить больше времени и проверять несколькими способами.

Но однозначно если у нас выборка насемплирована из дискретного распределения, то нельзя использовать критерий Колмогорова-Смирнова.

Если у нас данные категориальные или же количественные данные, имеющие дискретное распределения, то стоит воспользоваться критерием хи-квадрат.

Тест Колмогорова-Смирнова хороший, потому, что он чувствителен как к полодению так и к форме распределения, что для нас важно. 

Например, "традиционный метод" может показать равенство дисперсий и равенство мат ожиданий, но не сможет ничего сказать о различии форм.

Одно из интересных применений колмогорова-Смирнова на практике - при разбитие датасета на обучающую/тестовую выборки мы модем проверить будут ли они распределены из одного и того же распределения.

Чем ещё хорош Колмогоров-Смирнов: тем что распределение нормированной статистики не зависит от фнкций F и G. И если распределения выборок неизвестно или же не описывается каким-либо параметрическим семейством распределений, мы можем очень даже удачно его испольщзовать.

Но всё это в случае непрерывного, если же распределение дискретное, тогда можем использовать хи-квадрат.

**Резюме:**

Если мы понимаем, что выборки из непрерывных распределений, тогда будем использовать критерий Колмогорова-Смирнова.

Если же нет уверенности в этом, используем хи-квадрат, учитывая его ограничения на общее количество наблюдений и частоту встречи признаков наблюдаемых величин.

Другие приведенные критерии однородности могут лишь служить в качестве вспомогательной функции, если они выполняются, это нам не будет говорить о том, что точно две выборки из одного распределения, но если условия невыполнены, то это будет говорить о том, что принадлежности к одному распределению нет. В общем, это необходимость, но недостаточность.

Также 100% существует множемство других критериев однородности с которыми ещё предстоит познакомиться, которые будут лучше подходить к той или иной ситуации.
