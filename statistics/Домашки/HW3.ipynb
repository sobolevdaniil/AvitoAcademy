–û–±—ä—ë–º –≤—ã–±–æ—Ä–∫–∏ –≤ –ø–µ—Ä–≤–æ–º –∏ –≤—Ç–æ—Ä–æ–º —Å–ª—É—á–∞–µ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–π >=30, —Ç–∞–∫–∂–µ —á–∞—Å—Ç–æ—Ç–∞ –∫–∞–∂–¥–æ–π —è—á–µ–π–∫–∏ >=10, —á—Ç–æ –¥–∞–µ—Ç –Ω–∞–º –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏–π —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç –±–µ–∑ –ø–æ–ø—Ä–∞–≤–æ–∫.

–¢–∞–∫ –∫–∞–∫ —É –Ω–∞—Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ, –∏ —Ç–∞–∫ –∫–∞–∫ –º—ã —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –¥–≤–∞ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Ç–æ–≥–¥–∞ —á–∏—Å–ª–æ —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥—ã –±—É–¥–µ—Ç —Ä–∞–≤–Ω–æ (–∫–æ–ª-–≤–æ —Å—Ç–æ–ª–±—Ü–æ–≤ - 1)(–∫–æ–ª-–≤–æ —Å—Ç—Ä–æ–∫ - 1) = (2 - 1)(2 - 1) = 1

–ü—Ä–∏ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05 –ø—Ä–æ–≤–µ—Ä–∏–º —Å–ª–µ–¥—É—é—â—É—é –≥–∏–ø–æ—Ç–µ–∑—É:

$\textbf{H_0:}$ —Å—É—â–µ–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö —Ä–∞–∑–ª–∏—á–∏–π –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º –∏ –≤—Ç–æ—Ä—ã–º —Ä–∞–∑–æ–º –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è –∞–∫—Ü–∏–∏ –Ω–µ—Ç. CTR —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –Ω–µ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç —Ä–∞–∑–∞ –ø—Ä–æ–≤–µ–¥–µ–Ω–∏—è –∞–∫—Ü–∏–∏<br>
$\textbf{H_1:}$ —Ä–∞–∑–ª–∏—á–∏—è –µ—Å—Ç—å

–ü–æ—Å—Ç–æ—Ä–∏–º —Ç–∞–±–ª–∏—Ü—É —Å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º–∏ —á–∞—Å—Ç–æ—Ç–∞–º–∏:

````python
pd.DataFrame(
    [[68*38/136, 68*38/136, 38], [68*98/136, 68*98/136, 98], [68, 68, 136]], 
     index=['–ü–µ—Ä–≤—ã–π —Ä–∞–∑', '–í—Ç–æ—Ä–æ–π —Ä–∞–∑', '–í—Å–µ–≥–æ'], 
     columns=['–ù–µ –ø–µ—Ä–µ—à–ª–∏', '–ü–µ—Ä–µ—à–ª–∏', '–í—Å–µ–≥–æ'],
)
````

````python
import numpy as np
from collections import namedtuple
from scipy.stats import chi2
````

````python
#–ê–≤—Ç–æ–ø—Ä–æ–≤–µ—Ä–∫–∞

My–°hi2Results = namedtuple('My–°hi2Results', 
                                  ['is_rejected', 'pvalue'])


def my_chi2_for_contingency_tables(sample_A: list, sample_B: list, alpha: float = 0.05):
    """
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - sample_A: –ø–µ—Ä–≤–∞—è –≤—ã–±–æ—Ä–∫–∞
    - sample_B: –≤—Ç–æ—Ä–∞—è –≤—ã–±–æ—Ä–∫–∞
    - alpha: —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è.
        
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    - My–°hi2Results —Å –ø–æ–ª—è–º–∏:
        - is_rejected: bool
            - –æ—Ç–≤–µ—Ä–≥–ª–∞—Å—å –∏–ª–∏ –Ω–µ—Ç –≥–∏–ø–æ—Ç–µ–∑–∞ H_0 –Ω–∞ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha
        - pvalue: float
    """
    is_rejected = None
    pvalue = None
    
    k = 5 # –∫–æ—Ä–∑–∏–Ω—ã –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ sample_A
    m = 5 # –∫–æ—Ä–∑–∏–Ω—ã –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ sample_B
    
    # –†–∞–∑–æ–±—ä–µ–º sample_A –∏ sample_B –Ω–∞ k, m –∫–æ—Ä–∑–∏–Ω —Å–æ–æ—Ç–≤–µ—Ç—Å–≤–µ–Ω–Ω–æ, 
    # –¥–∏–∞–ø–∞–∑–æ–Ω—ã –∫–æ—Ç–æ—Ä—ã—Ö –±—É–¥—É—Ç —Ä–∞–≤–Ω—ã –º–µ–∂–¥—É —Å–æ–±–æ–π –∏ –ø—Ä–æ—Ç—è–≥–∏–≤–∞—Ç—å—Å—è –æ—Ç –º–∏–Ω –¥–æ –º–∞–∫—Å –∑–Ω–∞—á–µ–Ω–∏—è
    A = sorted(sample_A)
    B = sorted(sample_B)
    
    # —Å—á–∏—Ç–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω—ã –¥–ª—è sample_A
    step_A = (A[-1] - A[0]) / k
    delta_A = [A[0]]
    for i in range(1, k):
        delta_A.append(delta_A[-1] + step_A)
    delta_A.append(A[-1])
    
    # —Å—á–∏—Ç–∞–µ–º –¥–∏–∞–ø–∞–∑–æ–Ω—ã –¥–ª—è sample_B
    step_B = (B[-1] - B[0]) / m
    delta_B = [B[0]]
    for i in range(1, k):
        delta_B.append(delta_B[-1] + step_B)
    delta_B.append(B[-1])
    
    # —Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∫–∞–∂–¥—ã–π –∏–∑ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –¥–ª—è sample_A
    frequency_A = [0]*k
    for v in A:
        for ind, delta in enumerate(delta_A[1:]):
            if v <= delta:
                frequency_A[ind] += 1
                break
                
    # —Å—á–∏—Ç–∞–µ–º —á–∞—Å—Ç–æ—Ç—É –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∫–∞–∂–¥—ã–π –∏–∑ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –¥–ª—è sample_B            
    frequency_B = [0]*m
    for v in B:
        for ind, delta in enumerate(delta_B[1:]):
            if v <= delta:
                frequency_B[ind] += 1
                break
    
    AB = []
    for i in range(len(A)):
        AB.append((sample_A[i], sample_B[i]))
    
    frequencies = [[0] * m for i in range(k)]
    
    # —Å—á–∏—Ç–∞–µ–º —Å–æ–≤–º–µ—Å—Ç–Ω—É—é —á–∞—Å—Ç–æ—Ç—É –ø–æ–ø–∞–¥–∞–Ω–∏—è
    for a, b in AB:
        for ind, delta in enumerate(delta_A[1:]):
            if a <= delta:
                ind_A = ind
                break
        for ind, delta in enumerate(delta_B[1:]):
            if b <= delta:
                ind_B = ind
                break
        frequencies[ind_A][ind_B] += 1
    
    n = len(sample_A)
    
    # –°—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
    tau = 0
    for ind_A, line in enumerate(frequencies):
        for ind_B, v in enumerate(line):
            term = ((v - (frequency_A[ind_A]*frequency_B[ind_B])/n)**2) / (frequency_A[ind_A]*frequency_B[ind_B])
            tau += term
    tau *= n
    
    pvalue = 1 - chi2.cdf(x=tau, df=(m-1)*(k-1))
    if pvalue < alpha:
        is_rejected = True
    else:
        is_rejected = False

    return frequencies, My–°hi2Results(is_rejected, pvalue)
````

–ò –¥–∞–ª–µ–µ —Å—Ä–∞–≤–Ω–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–æ –∑–Ω–∞—á–µ–Ω–∏–µ–º —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç –¥–ª—è 2 —Å—Ç–µ–ø–µ–Ω–µ–π —Å–≤–æ–±–æ–¥, –ø—Ä–∏ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05

````python
from scipy.stats import chi2

1 - chi2.cdf(x=(52 - 2*91/3)**2 / (2*91/3) + (31 - 2*91/9)**2 / (2*91/9) + (8 - 91/9)**2 / (91/9), df=3 - 1)
````

**–í—ã–≤–æ–¥:**  pvalue < alpha = 0.05, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –º—ã —Å–∫–ª–æ–Ω—è–µ–º—Å—è –≤ —Å—Ç–æ—Ä–æ–Ω—É –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã: —Å–µ—Ä–≤–µ—Ä –ø–æ–ª—É—á–∞–µ—Ç –∑–∞–ø—Ä–æ—Å—ã —Å –∏–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é, –Ω–µ —Ä–∞–≤–Ω–æ–π 2/3. –û –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –≥–∏–ø–æ—Ç–µ–∑—ã  $ùêª_0$ –Ω–∏—á–µ–≥–æ —Å–∫–∞–∑–∞—Ç—å –Ω–µ–ª—å–∑—è.

## –ó–∞–¥–∞—á–∞ 4 (4 –±–∞–ª–ª–∞)


–ö–æ–º–ø–∞–Ω–∏—è –ø–æ –ø—Ä–æ–∏–∑–≤–æ–¥—Å—Ç–≤—É –∫–æ—Ä–æ–º–æ–≤ –¥–ª—è –ø–æ–ø—É–≥–∞–µ–≤ –ø—Ä–æ–≤–µ–ª–∞ –æ–ø—Ä–æ—Å —Å—Ä–µ–¥–∏ 5000 –≤–ª–∞–¥–µ–ª—å—Ü–µ–≤ –ø–æ–ø—É–≥–∞–µ–≤ –æ —Ç–æ–º, —á—Ç–æ –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞—é—Ç –µ—Å—Ç—å –∏—Ö –ø–∏—Ç–æ–º—Ü—ã. –ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –æ–ø—Ä–æ—Å–∞ –æ–∫–∞–∑–∞–ª–æ—Å—å, —á—Ç–æ 1010 –ø—Ç–∏—Ü –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞—é—Ç –≥–ª–∞–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Å—ã—Ä–∫–∏, 2200 &mdash; —Ñ—Ä—É–∫—Ç—ã, 950 &mdash; —Å–µ–º–µ—á–∫–∏ –∏ –∫—Ä—É–ø—ã, –∞ –≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞—é—Ç –∫–æ–ª–±–∞—Å—É. –ö–æ–º–ø–∞–Ω–∏—è —Å–¥–µ–ª–∞–ª–∞ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è —ç—Ç–∏—Ö –≤–∏–¥–æ–≤ –µ–¥—ã —Ä–∞–≤–Ω—ã $\theta, 2\theta, \theta, 1-4\theta$, –≥–¥–µ $\theta \in (0, 1/4)$. –í–µ—Ä–Ω–æ –ª–∏ –¥–∞–Ω–Ω–æ–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ?

**–ó–∞–¥–∞—á—É –Ω—É–∂–Ω–æ —Ä–µ—à–∏—Ç—å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏!**

–ú–æ–∂–Ω–æ –≤–æ—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –ª–∏—à—å –±–∏–±–ª–∏–æ—Ç–µ–∫–æ–π `from scipy.stats import chisquare`.

P.S. –ó–∞–¥–∞—á–∞ –ø–æ—è–≤–∏–ª–∞—Å—å –∏–∑ —Ä–∞—Ü–∏–æ–Ω–∞ –ø–æ–ø—É–≥–∞—è —Å–æ—Å—Ç–∞–≤–∏—Ç–µ–ª—è –¥–æ–º–∞—à–Ω–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è...

–ü—Ä–∏ —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05 –ø—Ä–æ–≤–µ—Ä–∏–º —Å–ª–µ–¥—É—é—â—É—é –≥–∏–ø–æ—Ç–µ–∑—É:

$\textbf{H_0:}$ –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è –≤–∏–¥–æ–≤ —Ä–∞–≤–Ω—ã $\theta, 2\theta, \theta, 1-4\theta$, –≥–¥–µ $\theta \in (0, 1/4)$<br>
$\textbf{H_1:}$ $H_0$ –Ω–µ–≤–µ—Ä–Ω–∞

````python
pd.DataFrame(
    [[1010, 2200, 950, 840], ['$\theta$', '$2\theta$', '$\theta$', '1-4*$\theta$']], 
     index=['–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —á–∞—Å—Ç–æ—Ç–∞', '–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —á–∞—Å—Ç–æ—Ç–∞ / 5000'], 
     columns=['–ì–ª–∞–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Å—ã—Ä–∫–∏', '–§—Ä—É–∫—Ç—ã', '–°–µ–º–µ—á–∫–∏ –∏ –∫—Ä—É–ø—ã', '–ö–æ–ª–±–∞—Å–∞'],
)
````

–î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –∫—Ä–∏—Ç–µ—Ä–∏–π —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–±–æ—Ç–∞–ª –Ω–∞ –∑–∞–¥–∞—á–µ —Å –æ—Ü–µ–Ω–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –Ω–∞–º –Ω—É–∂–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è.

$P = \theta^{1010} * (2\theta)^{2200} * \theta^{950} * (1-4\theta)^{840}$

$ln(P) = ln(\theta)*1010 + ln(2)*2200 + ln(\theta)* 2200 + ln(\theta)*950 + ln(1-4\theta)*840$

–î–∞–ª–µ–µ –≤–æ–∑—å–º–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—É—é –∏ –ø—Ä–∏—Ä–∞–≤–Ω—è–µ–º –∫ –Ω—É–ª—é:

$(1010 + 2200 + 950) * (1/\theta) + 840 * (1/(1-4\theta))*(-4) = 0$

$4160 * (1-4\theta) = 3360 * \theta$

–ü–æ–ª—É—á–∏–º $\theta = 4160 / (3360 + 4160*4) = 0.208$ –î–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª.

–¢–∞–∫ –∫–∞–∫ –º—ã –æ—Ü–µ–Ω–∏–ª–∏ –æ–¥–∏–Ω –ø–∞—Ä–∞–º–µ—Ç—Ä, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –±—É–¥–µ–º —Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å–æ —Å—Ç–µ–ø–µ–Ω—è–º–∏ —Å–≤–æ–±–æ–¥: (4 - 1 - 1) (ddof=1)

````python
from scipy.stats import chisquare

chisquare([1010, 2200, 950, 840], f_exp=[0.208*5000, 2*0.208*5000, 0.208*5000, 0.168*5000], ddof=1)
````

**–í—ã–≤–æ–¥:** pvalue < alpha = 0.05, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –æ—Å–Ω–æ–≤–∞–Ω–∏—è –ø—Ä–∏–Ω—è—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –Ω–µ—Ç, —Å–∫–ª–æ–Ω—è–µ–º—Å—è –≤ —Å—Ç–æ—Ä–æ–Ω—É –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π.
