непонятный скачок стандартного отклонения, по-хорошему нужно выяснить почему мы это увидели на графике

```python
df_basic_amount = df[df['product'] == 'basic']['amount']
df_basic_transactions = df[df['product'] == 'basic']['transactions']

df_advanced_amount = df[df['product'] == 'advanced']['amount']
df_advanced_transactions = df[df['product'] == 'advanced']['transactions']
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

def decompose(df_for_dec):
    for df in df_for_dec:
        decomposed_1 = seasonal_decompose(df, period=100_000)
        
        plt.figure(figsize=(12, 16))

        plt.subplot(311)
        decomposed_1.trend.plot(ax=plt.gca())
        plt.title('Trend')

        plt.subplot(312)
        decomposed_1.seasonal.plot(ax=plt.gca())
        plt.title('Seasonality')

        plt.subplot(313)
        decomposed_1.resid.plot(ax=plt.gca())
        plt.title('Residuals')

        plt.tight_layout()
        
decompose([df_basic_amount, df_basic_transactions, df_advanced_amount, df_advanced_transactions])
```

Сезональность некая есть, но её нужно бы наблюдать в более мелких границах времени.

Остатки необъяснимы.

- Тренд basic_amount топчится на месте 
- Тенд basic_transactions постепенно шел вверх, был большой скачок 2020-12-01
- Тренд advanced_amount линейно идет наверх
- Тренд advanced_transactions шел наверх до 2020-11-15, далее неизменялся
